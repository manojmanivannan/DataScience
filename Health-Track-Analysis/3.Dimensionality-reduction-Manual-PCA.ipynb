# Dimensionality-reduction PCA Manual

```python
import pandas as pd
import numpy as np
#use widget instead of inline to make the plot interactive
%matplotlib inline 
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
# plt.rcParams['figure.figsize'] = [18, 11]
# font = {'size'   : 22}

plt.rc('font', **font)
# plt.rcParams['figure.figsize'] = [15, 10]
ht_grouped = pd.read_csv('data/ht_grouped_feature_engineered.csv',index_col='device_id')
# https://towardsdatascience.com/a-complete-guide-to-principal-component-analysis-pca-in-machine-learning-664f34fc3e5a
```

```python
ht_grouped.head()
```

```python
feature_len = ht_grouped.drop('lifestyle',axis=1).columns.values.__len__()
```

```python
X = ht_grouped.drop('lifestyle',axis=1).values
y = ht_grouped.lifestyle.values
```

```python
from sklearn.preprocessing import StandardScaler 
X_std = StandardScaler().fit_transform(X)
```

```python
#mean_vec = np.mean(X_std, axis=0) 
#cov_mat = (X_std - mean_vec).T.dot((X_std - mean_vec)) / (X_std.shape[0]-1) 
#print('Covariance matrix \n%s' %cov_mat) 
print('Covariance matrix \n') 
cov_mat= np.cov(X_std, rowvar=False) 
cov_mat
```

```python
cov_mat = np.cov(X_std.T) 
eig_vals, eig_vecs = np.linalg.eig(cov_mat) 
print('Eigenvectors \n%s' %eig_vecs) 
print('\nEigenvalues \n%s' %eig_vals)
```

```python
sq_eig=[] 
for i in eig_vecs: 
    sq_eig.append(i**2) 
    print(sq_eig) 
sum(sq_eig) 
print("sum of squares of each values in an eigen vector is \n", 2.23466464e-01+ 1.54732629e-02+ 2.03004679e-05+ 5.12992101e-01+ 7.13026850e-03+ 8.52287437e-02+ 1.24049769e-01+ 1.09474404e-04+3.07172181e-02+ 6.10117336e-04+ 2.02280577e-04) 
for ev in eig_vecs: np.testing.assert_array_almost_equal(1.0, np.linalg.norm(ev),decimal=3)
```

```python
#Make a list of (eigenvalue, eigenvector) tuples 
eig_pairs = [(np.abs(eig_vals[i]), eig_vecs[:,i]) for i in range(len(eig_vals))]
```

```python
#Sort the (eigenvalue, eigenvector) tuples from high to low 
eig_pairs.sort() 
eig_pairs.reverse() 
# print("\n",eig_pairs) 
#Visually confirm that the list is correctly sorted by decreasing eigenvalues 
print('\n\n\nEigenvalues in descending order:') 
for i in eig_pairs: 
    print(i[0])
```

```python
num_k = 5
tot = sum(eig_vals) 
print("\n",tot) 
var_exp = [(i / tot)*100 for i in sorted(eig_vals, reverse=True)] 
print("\n\n1. Variance Explained\n",var_exp) 
cum_var_exp = np.cumsum(var_exp) 
print("\n\n2. Cumulative Variance Explained\n",cum_var_exp) 
print("\n\n3. Percentage of variance the first two principal components each contain\n ",var_exp[0:num_k]) 
print("\n\n4. Percentage of variance the first two principal components together contain\n",sum(var_exp[0:num_k]))
```

```python
# print(eig_pairs[0][1]) 
# print(eig_pairs[1][1])
# print(eig_pairs[2][1])
matrix_w = np.hstack((eig_pairs[0][1].reshape(feature_len,1), 
                      eig_pairs[1][1].reshape(feature_len,1),
                      eig_pairs[2][1].reshape(feature_len,1),
                      eig_pairs[3][1].reshape(feature_len,1),
                      eig_pairs[4][1].reshape(feature_len,1),
                      eig_pairs[5][1].reshape(feature_len,1),)) 
#hstack: Stacks arrays in sequence horizontally (column wise). 
print('Matrix W:\n', matrix_w)
matrix_w[0]
```

```python
Y = X_std.dot(matrix_w) 
principalDf = pd.DataFrame(data = Y , columns = ['principal component {}'.format(s+1) for s in range(len(matrix_w[0]))]) 
principalDf.head()
```

```python
finalDf = pd.concat([principalDf,pd.DataFrame(y,columns = ['lifestyle'])], axis = 1) 
finalDf.head()
```

```python
finalDf.to_pickle('data/dim_reduced_df.pkl')
```

```python
fig = plt.figure(figsize = (14,8)) 
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15) 
ax.set_ylabel('Principal Component 2', fontsize = 15) 
ax.set_title('2 Component PCA', fontsize = 20) 
targets = ['Sedentary','Weight Trainer','Athlete','Cardio Enthusiast']
colors = ['r', 'g', 'b','c'] 
for target, color in zip(targets,colors): 
    indicesToKeep = finalDf['lifestyle'] == target  
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1'] , 
               finalDf.loc[indicesToKeep, 'principal component 2'] , 
               c = color , s = 50) 
ax.legend(targets) 
ax.grid()
```
