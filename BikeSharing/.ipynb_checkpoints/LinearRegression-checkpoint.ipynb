```python
import pandas as pd
import numpy as np
#use widget instead of inline to make the plot interactive
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
```

```python
df = pd.read_pickle('data/df_day_pc.pkl')
X_features = df[['principal component {}'.format(s+1) for s in range(9)]]
y_target = df['casual']
```

```python
df.head()
```

```python
from tpot import TPOTRegressor
# importing train_test_split from sklearn
from sklearn.model_selection import train_test_split
# splitting the data
x_train, x_test, y_train, y_test = train_test_split(X_features, y_target, test_size = 0.15, random_state = 42)
```

```python
tpot = TPOTRegressor(generations=5, population_size=100,verbosity=2,n_jobs=-1)
```

```python
tpot.fit(x_train, y_train)
```

```python
print(tpot.score(x_test,y_test))
```

```python
df_day = pd.read_pickle('data/df_day.pkl')
```

```python
X_features = df_day.drop(['instant','casual','registered','cnt'],axis=1)
y_target = df_day['casual']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X_features, y_target, test_size = 0.15, random_state = 42)
```

```python
tpot.fit(x_train, y_train)
print(tpot.score(x_test,y_test))
```
